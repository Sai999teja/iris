```python
from sklearn.model_selection import train_test_split
```

```python
x = iris.drop('species',axis =1)
```

```python
y = iris['species']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 100,test_size =0.30)
```

```python
x_train.head()
```

# model building

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.ensemble import RandomForestClassifier
```

```python
ranclf = RandomForestClassifier(min_samples_split = 10,max_depth = 15)
```

```python
model = ranclf.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = model.predict(x_test)
```

```python
y_train_pred = model.predict(x_train)
```

```python
accuracy_score(y_train_pred,y_train)
```

```python
accuracy_score(y_pred,y_test)
```

# by simple tuning of parameters we can reach upto 98% accuracy
